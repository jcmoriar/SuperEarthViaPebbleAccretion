
#Super-Earth Formation via Pebble Accretion
###General idea

Pebble accretion is shown to be a good mechanism to quickly accrete material in the outer disk (as a means to solve the core formation timescale problem). There is no reason this mechanism wouldn’t operate at smaller semi-major axes with some difference in efficiency. Additionally, the idea fits in naturally with the notion that planets are forming within an accretion disk.

###Outline
####Pebble production: 
The flux of pebbles through the disk is set by the pebble production rate in the outer parts of the disk. This is because the growth timescales are much longer than the migration timescales (100K vs 1K years). I will use the pebble production rates derived in Lambrechts and Johansen 2014. The idea behind this is that one sets the growth timescale equal to the migration timescale. This defines a radial front in the disk where pebbles are growing large enough to begin migrating. The expansion of this front through the disk sets the pebble production rate. One thing to consider that LJ14 did not is the time evolution of the disk. This would decrease the pebble production at later times. Ignoring that for the time being the derived pebble production rate is

$$F = 9.5\times 10^{-5} \left(\frac{\beta}{500\, g\, cm^{-2}}\right)\left(\frac{M_*}{M_{\odot}}\right)^{1/3}\left(\frac{Z_0}{0.01_{ }}\right)^{5/3}\left(\frac{t}{10^6yr}\right)^{-1/3} M_E\, yr^{-1}$$
   where $\beta$ is the initial gas surface density at 1AU and $Z_0$ is the initial dust to gas ratio.   
   
   
   
####Pebble surface density: 
The pebble surface density can be derived (as in LJ14) using the continuity equation. The flux of pebbles through the disk,

$$F=2\pi r v_r \Sigma_p.$$

F is set by the pebble production rate. Once we know the drift velocity of the pebbles throughout the disk we can solve for the surface density. In LJ14 this was done by balancing the drift timescale with the pebble coagulation timescale (similar to how the production rate was derived). That work was mostly concerned with the outer parts of the disk where predicted pebble sizes were small and sticking efficiencies between pebbles high. This work focuses on the inner 1 AU, where pebble sizes predicted from the LJ14 derivation would be high. The assumption of efficient sticking at this point is no longer valid. If we look at more sophisticated coagulation simulations (e.g. Birnstiel et al. 2010), we see that pebbles typically don’t grow larger than cm size. 

With the assumption that the dominant pebble size throughout the inner disk is a constant $a_0$ let’s re-derive the pebble surface density. We begin by dertermining the drift velocity of particles,




$$v_r = -2\frac{St}{St^2+1}\eta v_K \approx -2St\,\eta v_K \;\;\;(\mathrm{for}\; St \lesssim 1) $$

where $St$ is the Stokes number, $v_K$ the Keplerian velocity and $\eta$ is a measure of radial gas support given by (eqn 18 of LJ14),

$$\eta = 0.0015\left(\frac{r}{\mathrm{AU}}\right)^{1/2}.$$

The Stokes number is given as,

$$St = \frac{\rho a}{\rho_g c_s} \;\;\; a<\frac{9\lambda}{4}$$

$$\;\;\;\;\;\;\;= \frac{4\rho a^2}{9\rho_g c_s \lambda} \;\;\; a > \frac{9\lambda}{4}$$

where $\rho$ is the material density of the pebbles, $a$ is their size, $\rho_g$ is the gas density, $c_s$ is the isothermal sound spped and $\lambda$ is the mean free path of the gas. Setting $a=a_0$ and combining the continuity and radial drift equations gives,

$$\Sigma_p = \frac{F}{ 4 \pi r \eta v_k St(a_0)}$$ 

####Protoplanet Growth

The rate at which a protoplanet grows can be calculated using its collisional cross section as,
$$\frac{dM}{dt} = \sigma_c \rho_p v_{rel}, $$


where $\rho_p$ is the volume density of accreting particles and $v_{rel}$ is the relative velocity between the protoplanet and the accreting particles. In terms of the impact parameter, $b$, this can be written as,
$$\frac{dM}{dt} = \pi b^2 \rho_p v_{rel}.$$

In the case where the impact parameter is larger than the scale height of the disk (so called 2-D accretion),
$$\frac{dM}{dt} = 2b\Sigma_p v_{rel},$$

where $\Sigma_p$ is the surface density of accreting particles in the disk. Ormel and Klahr (2010) carried out numerical simulations of the accretion of pebbles including the influence of gas drag. They determined the impact parameter as a function of particle/protoplanet size and disk properties. See their paper for details. I will use their calculated impact parameters in my calculations of planet growth. The relative velocity between the particles and the embryo depends on the conditions. We take $v_{rel}$ to be the maximum of the drift velocity, turbulent velocity, shear velocity and relative velocity due to an eccentric orbit where,
$$v_{turb}^2 = \alpha c_s^2 \times min\left( 2St, \frac{1}{1+St}\right) $$
$$v_{drift} \approx \eta v_{kep} $$
$$v_{ecc} = ev_{kep}$$
$$v_{shear} = \frac{bv_{kep}}{r}.$$

#### Growth of planetesimals
In this work we will consider planetesimals to be bodies that are small and numerous enough that they can be treated in a statistical sense. The importance of planetsimals in this model is that they are another way to halt the inward drift of pebbles and keep mass at larger semi-major axes. The amount of mass that planetesimals compared to embryos trap is hard predict because we do not know the mass function or birth rate of planetesimals. It is also unknown how important the mass function is in determining the rate at which pebbles are swept up. Luckily, this is not too difficult to determine using the pebble accretion rates from the above equations.

#### Overall Model
The overall model has three components: pebbles, planetesimals and embryos. The growth of embryos is followed through a combination of n-body simulations and calculated pebble and planetesimal accretion rates. The planetesimal disk will deplete through collisions with embryos and grow from accretion of and formation from pebbles. The pebble disk mass is governed by the pebble production rate already outlined and the accretion of pebbles onto planetesimals and embryos.

For the planetesimals accretion/formation rate I'll just assume most of the mass comes from accreting pebbles not from direct formation of planetesimals from pebbles. From there, I assume that the slope of the planetesimal mass function doesn't change. I guess this assentially means that the planetesimal disk mass will grow up to a steady state where all new pebble accretion is essentially transfered in to an embryo. So the point is: what is the rate of pebble accretion? Integrated over all planetesimals? How do we take into account that they are competing for pebbles?

####Importance of planetesimal mass function
I will start with the simplest comparison. Given a set amount of mass in planetesimals all of size R, how does the pebble accretion rate depend on R? The following plot shows that the total pebble accretion rate of a disk of planetesimals is about the same (to a factor of 10) regardless of planetesimal size for sizes in the range 10-1000 km in radius. However, if the planetsimals are small $\sim$1km in radius, the disk is more effective at accreting pebbles. Additionally, larger embryos behave differently, accreting less mass at small semi-major axes and more mass at larger semi major axes. This is a fortunate result as planetesimals are generally believed to be in this 10-1000 km size range and therefore I don't have to worry about the size distribution of planetesimals. 

```julia
include("../src/pebble_accretion.jl")
using Gadfly
set_default_plot_size(25cm, 15cm)
m_tot = 0.05
time = 1e5
a1(x) = total_accretion_rate(10^x, time, m_tot, body_size=1.)
a10(x) = total_accretion_rate(10^x, time, m_tot, body_size=10.)
a100(x) = total_accretion_rate(10^x, time, m_tot, body_size=100.)
a1000(x) = total_accretion_rate(10^x, time, m_tot, body_size=1000.)
a10000(x) = total_accretion_rate(10^x, time, m_tot, body_size=10000.)
plot([a1, a10, a100, a1000, a10000], 5e-2, 1e0, Scale.x_log10, Scale.y_log10, 
Guide.xlabel("Semi-major axis [AU]"), Guide.ylabel("Pebble accretion rate [M_Earth/yr]"),
color=["1", "10", "100", "1000", "10000"], Guide.colorkey("Size [km]"), Theme(line_width=2pt))
```

**Note:** I think the reason why this is occuring is because gas drag starts becoming effective for bodies 100km and larger in this part of the disk. If there were no gas drag, we would expect that small planetesimals would accrete more mass as they have a larger collective cross-section. However, as we grow in size, gas drag begins to counteract the decrease in physical cross section and thus we get similar combined accretion rates. We can investigate this further by calculating the impact parameter and normalizing to the physical size of the planetesimal.
