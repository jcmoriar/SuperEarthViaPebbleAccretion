```julia
include("../src/pebble_accretion.jl")
time = 1e5
r0 = 1e0
planet_size = [10^x for x=linspace(0, 4, 100)]
a = [10^x for x=linspace(1, -1, 100)]
rmin = reshape([r for r=planet_size[1:end-1], sma=a[1:end-1]], (length(planet_size)-1)*(length(a)-1))
rmax = reshape([r for r=planet_size[2:end], sma=a[1:end-1]], (length(planet_size)-1)*(length(a)-1))
smamin = reshape([sma for r=planet_size[1:end-1], sma=a[1:end-1]], (length(planet_size)-1)*(length(a)-1))
smamax = reshape([sma for r=planet_size[2:end], sma=a[2:end]], (length(planet_size)-1)*(length(a)-1))

# data[:AccretionTimeScale] = [growth_rate(data[:SemiMajorAxis][i], time, data[:PlanetSize][i], r0) for i=1:length(data[:PlanetSize])]
τ = reshape([log10(PA.impact_parameter(smamin[i], time, rmin[i], r0)/1000./rmin[i]) for i=1:length(smamin)], length(a)-1, length(planet_size)-1);

τ = transpose(τ);
τ[1,1]=0
planet_size=log10(planet_size)
a = log10(a);
```
