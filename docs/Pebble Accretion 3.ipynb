###Statistic
We basically need some statistic to compare to the Kepler systems. Rory mentioned the total solid mass which might be a good and simple one. We can then vary a few of the input parameters to see what the resulting final mass is. A strong result would show that pebble accretion is robust. That is, it depends weakly on the input parameters.

```julia
include("../src/pebble_accretion.jl")
PA.set_globals(pebble_density = 1e0,
               α_t = 1e-3,
               pebble_size = 1e0,
               min_protoplanet_size = 1e2,
               max_protoplanet_size = 5e2,
               protoplanet_density = 3e0,
               β_0 = 5e2,
               τ_dissipation = 3e6,
               dust_to_gas = 1e-2,
               initial_surface_density = 1e-2,
               α_slope = -1.5 )
d = PA.Disk(0.05, 5.);
f = open("../data/Run01.dat", "w")

f2 = open("../data/Run01_flux.dat", "w")
PA.writeline_inwardflux(d, 0., f2, num_bins=100, header=true)
PA.writeline_inwardflux(d, 0., f2, num_bins=100)


PA.write_header(f)
PA.writeline_surfacedensity(d, 0., f, num_bins=100)
close(f)
close(f2)
f = open("../data/Run01.dat", "a")
f2 = open("../data/Run01_flux.dat", "a")


t=1e5
dt=1e3
pm = 0
am = 0
for i=1:5
    print("\r $i")
    a, p = PA.accrete_pebbles!(d, t, dt)
    pm += p
    am += a
    PA.writeline_surfacedensity(d, t, f, num_bins=100)
    PA.writeline_inwardflux(d, t, f2, num_bins=100)
    t += dt
end
close(f)
close(f2)
```
